# First point

```python
from sklearn.svm import SVC
clf = SVC(kernel = "linear")
clf.fit(X, y)
y_predicted = clf.predict(X)
print(y_predicted)
```

# Second point

```python
# get support vectors
support_vectors = clf.support_vectors_
print("Number of support vectors: ", len(support_vectors))
print("First class: ", clf.n_support_[0], "Second class: ", clf.n_support_[1])

plt.scatter(X[:,0], X[:,1], c = y)

# https://stackoverflow.com/questions/23794277/extract-decision-boundary-with-scikit-learn-linear-svm
# http://scikit-learn.org/stable/auto_examples/svm/plot_svm_margin.html
# http://scikit-learn.org/stable/auto_examples/svm/plot_separating_hyperplane.html#sphx-glr-auto-examples-svm-plot-separating-hyperplane-py
# plot the decision function
# xlim = ax.get_xlim()
# ylim = ax.get_ylim()
w = clf.coef_[0]
a = -w[0] / w[1]

xx = xx = np.linspace(-2, 2)
yy = a * xx - (clf.intercept_[0]) / w[1]

margin = 1 / np.sqrt(np.sum(clf.coef_ ** 2))
yy_down = yy - np.sqrt(1 + a ** 2) * margin
yy_up = yy + np.sqrt(1 + a ** 2) * margin

plt.plot(xx, yy, 'k-')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')

# plot support vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80, zorder=10, edgecolors='k')
plt.show()
```

```python
a = np.array([xx[0], yy_down[0]])
b = np.array([xx[29], yy_down[29]])

c = np.array([xx[0], yy_up[0]])
d = np.array([xx[29], yy_up[29]])

e = np.array([xx[0], yy[0]])
f = np.array([xx[29], yy[29]])

margin_error = 0

for i in range(len(X[:, 0])):
    p = np.array([X[i, 0], X[i, 1]])
    #val1 = (clf.support_vectors_[i, 0] - A[0])*(B[1] - A[1]) - (clf.support_vectors_[i, 1] - A[1])*(B[0] - A[0])
    #val2 = (clf.support_vectors_[i, 0] - C[0])*(D[1] - C[1]) - (clf.support_vectors_[i, 1] - C[1])*(D[0] - C[0])
    #val3 = (clf.support_vectors_[i, 0] - E[0])*(F[1] - E[1]) - (clf.support_vectors_[i, 1] - E[1])*(F[0] - E[0])
    val1 = np.cross(p-a, b-a) < 0 # margine basso
    val2 = np.cross(p-c, d-c) < 0# margine alto
    #print(val1,val2)
    if val1 == np.True_ and val2 == np.False_:
        margin_error = margin_error + 1

print("Margin error: ", margin_error)
```

```python
def classification_error(y_predicted, y_true): 
    wrong = 0
    for i in range(len(y_predicted)):
        if(y_predicted[i] != y_true[i]):
            wrong = wrong + 1
    return wrong

print("Number of wrongly classified example: ", classification_error(y_predicted, y))
```
