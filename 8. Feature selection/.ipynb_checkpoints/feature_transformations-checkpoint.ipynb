### Loading dataset

```python
import numpy as np
from __future__ import division

np.set_printoptions(threshold=np.nan)

f = open("australian.data")
dataset =  np.loadtxt(fname = f, delimiter = ',')
print(dataset)
print(dataset.shape)
```

```python
# Valore 0 è positivo CLASSE 1
# Valore 1 è negativo CLASSE 2
```

### Info and split

```python
n_elements = len(dataset[:, 0])
print("Numero elementi dataset: " + str(n_elements))
n_features = len(dataset[0, :])
print("Numero features dataset: " + str(n_features))

true_class = dataset[:, n_features - 1]
data = np.delete(dataset, n_features - 1, axis = 1)

n_features = n_features - 1

print(dataset.shape)
print(true_class.shape)
print(data.shape)
```

### Split train and test example

```python
from sklearn.model_selection import train_test_split

X = data
y = true_class

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Without Pre-processing

```python
from sklearn.metrics import (brier_score_loss, precision_score, recall_score, f1_score, accuracy_score)
```

```python
def compute_metric(y_test, y_pred):
    print("Precision: %1.3f" % precision_score(y_test, y_pred))
    print("Recall: %1.3f" % recall_score(y_test, y_pred))
    print("F1: %1.3f" % f1_score(y_test, y_pred))
    print("Accuracy: %1.3f\n" % accuracy_score(y_test, y_pred))
```

## Gaussian model

```python
from sklearn.naive_bayes import GaussianNB

clf_gaussianNB = GaussianNB()
clf_gaussianNB.fit(X_train, y_train)
proba_clf_gaussianNB = clf_gaussianNB.predict_proba(X_test)
predicted_clf_gaussianNB = clf_gaussianNB.predict(X_test)

compute_metric(y_test, predicted_clf_gaussianNB)
```

## Logistic regression

```python
from sklearn import linear_model

clf_logisticR = linear_model.LogisticRegression(C = 1e5)
clf_logisticR.fit(X_train, y_train)
proba_clf_logisticR = clf_logisticR.predict_proba(X_test)
predicted_clf_logisticR = clf_logisticR.predict(X_test)

compute_metric(y_test, predicted_clf_logisticR)
```

## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

clf_decisionTree = DecisionTreeClassifier(random_state=42)
clf_decisionTree.fit(X_train, y_train)
proba_clf_decisionTree = clf_decisionTree.predict_proba(X_test)
predicted_clf_decisionTree = clf_decisionTree.predict(X_test)

compute_metric(y_test, predicted_clf_decisionTree)
```

# With Pre-processing

Si potrebbe anche pensare di fare una cosa diversa: dividere il dataset in due sotto-gruppi, uno contentente i dati categorici, il secondo quelli continui. Utilizzare due algoritmi differenti, uno per ogni sotto-gruppo appena creato. Confrontare con un terzo classificatore le probabilità con cui una certa classe è stata predetta da ogni classificatore (i due di appena citati) per ogni elemento del dataset (i 690 esempi) e trovare il valore finale predetto. Insomma, ensemble learning.

```python
from sklearn import decomposition

def compute_pca(dataX, n_components_):
    pca = decomposition.PCA(n_components = n_components_)
    pca.fit(dataX)
    newDataX = pca.transform(dataX)
    return newDataX
```

## Apply calibration

```python
# https://en.wikipedia.org/wiki/Naive_Bayes_classifier#Examples
# http://scikit-learn.org/stable/modules/naive_bayes.html
# http://scikit-learn.org/stable/modules/preprocessing.html#encoding-categorical-features

index_catego = [0, 3, 4, 5, 7, 8, 10, 11]

dataset_calibration = dataset

output_positive = dataset_calibration[:, index_catego][dataset_calibration[:, 14] == 0]
mean_p = output_positive.mean(axis = 0)
var_p = output_positive.var(axis = 0)

output_negative = dataset_calibration[:, index_catego][dataset_calibration[:, 14] == 1]
mean_n = output_negative.mean(axis = 0)
var_n = output_negative.var(axis = 0)

n_positive = (dataset_calibration[:, 14] == 0).sum()
n_negative = (dataset_calibration[:, 14] == 1).sum()

prob_positive = n_positive / (n_positive + n_negative)
prob_negative = n_negative / (n_positive + n_negative)
```

```python
for i in range(dataset_calibration.shape[0]):
    if dataset_calibration[i, 14] == 0:
        j = 0
        for k in index_catego:
            prob_temp_1 = (1 / (np.sqrt(2 * np.pi * (var_p[j] ** 2)))) * np.exp((-(dataset_calibration[i, k] - mean_p[j]) ** 2) / (2 * (var_p[j] ** 2)))
            prob_temp_2 = (1 / (np.sqrt(2 * np.pi * (var_n[j] ** 2)))) * np.exp((-(dataset_calibration[i, k] - mean_n[j]) ** 2) / (2 * (var_p[j] ** 2)))
            dataset_calibration[i, k] = (prob_positive * prob_temp_1) / (prob_positive * prob_temp_1 + prob_negative * prob_temp_2)
            j = j + 1
    if dataset_calibration[i, 14] == 1:
        p = 0
        for k in index_catego:
            prob_temp_1 = (1 / (np.sqrt(2 * np.pi * (var_p[p] ** 2)))) * np.exp((-(dataset_calibration[i, k] - mean_p[p]) ** 2) / (2 * (var_p[p] ** 2)))
            prob_temp_2 = (1 / (np.sqrt(2 * np.pi * (var_n[p] ** 2)))) * np.exp((-(dataset_calibration[i, k] - mean_n[p]) ** 2) / (2 * (var_p[p] ** 2)))
            dataset_calibration[i, k] = (prob_negative * prob_temp_2) / (prob_positive * prob_temp_1 + prob_negative * prob_temp_2)
            p = p + 1
print(dataset_calibration)
```

```python
data_calibration = np.delete(dataset_calibration, n_features - 1, axis = 1)

X_calibration = data_calibration
X_train_calibration, X_test_calibration, y_train_calibration, y_test_calibration = train_test_split(X_calibration, y, test_size=0.33, random_state=42)
```

```python
# http://www.statstutor.ac.uk/resources/uploaded/paired-t-test.pdf
# https://towardsdatascience.com/inferential-statistics-series-t-test-using-numpy-2718f8f9bf2f
# https://machinelearningmastery.com/how-to-code-the-students-t-test-from-scratch-in-python/

from scipy.stats import t

degree_of_freedom = n_fold - 1
alpha = 0.05
cv = t.ppf(1.0 - alpha, degree_of_freedom)

paired_t_test = np.zeros(shape=(15, 3))
print(paired_t_test.shape)
for i in range(len(paired_t_test) - 5):
    paired_t_test[i, 0] = clf_gaussianNB_kfold_discretization_scores[i] - clf_logisticR_kfold_scores[i]
    paired_t_test[i, 1] =  clf_gaussianNB_kfold_discretization_scores[i] - clf_decisionTree_scores[i]
    paired_t_test[i, 2] = clf_logisticR_kfold_scores[i] - clf_decisionTree_scores[i]

# Mean of each difference
paired_t_test[10, :] = [np.mean(paired_t_test[0:10, 0]), np.mean(paired_t_test[0:10, 1]), np.mean(paired_t_test[0:10, 2])]
# Standard deviation of each difference
paired_t_test[11, :] = [np.std(paired_t_test[0:10, 0]), np.std(paired_t_test[0:10, 1]), np.std(paired_t_test[0:10, 2])]
# Standard error of each difference
paired_t_test[12, :] = paired_t_test[11, :] / np.sqrt(n_fold)
# T value of each difference
paired_t_test[13, :] = paired_t_test[10, :] / paired_t_test[12, :]
# P value of each difference
paired_t_test[14, :] = [1 - stats.t.cdf(paired_t_test[13, 0], df = degree_of_freedom), 1 - stats.t.cdf(paired_t_test[13, 1], df = degree_of_freedom), 1 - stats.t.cdf(paired_t_test[13, 2], df = degree_of_freedom)]
paired_t_test[14, :] = paired_t_test[14, :] * 2

print(paired_t_test)
```

```python
# https://machinelearningmastery.com/nonparametric-statistical-significance-tests-in-python/
# https://stats.stackexchange.com/questions/306841/z-score-on-wilcoxon-signed-ranks-test

from scipy import stats

print(stats.wilcoxon(paired_t_test[0:10, 0], zero_method = "wilcox"))
print(stats.wilcoxon(paired_t_test[0:10, 1], zero_method = "wilcox"))
print(stats.wilcoxon(paired_t_test[0:10, 2], zero_method = "wilcox"))
```

```python
from numpy import compress

def compute_wilcoxon(difference_score):
    data = compress(np.not_equal(difference_score, 0), difference_score, axis = -1)
    count = len(data)
    rank = stats.rankdata(abs(data))
    rank_plus = np.sum((data > 0) * rank, axis = 0)
    rank_minus = np.sum((data < 0) * rank, axis = 0)
    T = min(rank_plus, rank_minus)
    mean_W = (count * (count + 1.)) / 4 # normal approximation
    var_W = count * (count + 1.) * (2. * count + 1.) # normal approximation
    replist, repnum = stats.find_repeats(rank)
    if repnum.size != 0:
        var_W -= (pow(repnum.size * 2, 3) - repnum.size * 2) / 48
    var_W = np.sqrt(var_W / 24)
    z = (rank_plus - mean_W) / var_W
    p_value = 2. * stats.norm.sf(abs(z))
    return p_value
```

```python
print(compute_wilcoxon(paired_t_test[0:10, 0]))
print(compute_wilcoxon(paired_t_test[0:10, 1]))
print(compute_wilcoxon(paired_t_test[0:10, 2]))
```
