# # Classifiers comparison: decision trees and k-nearest neighbors on the dataset Boston


In the following program we compare the prediction results obtained by decision trees and k-nearest neighbors on the dataset Boston

```python
from sklearn import tree 
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

boston = load_boston()
X = boston.data
y = boston.target
print(X.shape)
print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
clf_tree = tree.DecisionTreeRegressor(random_state=0)
clf_tree.fit(X_train, y_train)

predicted_tree = clf_tree.predict(X_test)

print("RMS Tree: %r " % np.sqrt(np.mean((predicted_tree - y_test) ** 2)))
```

```python
from sklearn import neighbors
n_neighbors = 4
clf_knn_uniform = neighbors.KNeighborsRegressor(n_neighbors, weights = 'uniform')
clf_knn_uniform.fit(X_train, y_train)

predicted_knn_uniform = clf_knn_uniform.predict(X_test)

print("RMS Knn uniform: %r " % np.sqrt(np.mean((predicted_knn_uniform - y_test) ** 2)))
```

```python
clf_knn_distance = neighbors.KNeighborsRegressor(n_neighbors, weights = 'distance')
clf_knn_distance.fit(X_train, y_train)

predicted_knn_distance = clf_knn_distance.predict(X_test)

print("RMS Knn distance: %r " % np.sqrt(np.mean((predicted_knn_distance - y_test) ** 2)))
```

```python
# Plot the results
plt.figure()
plt.scatter(y_test, predicted_tree, s=20, edgecolor="black",c="darkorange", label="Tree data")
plt.scatter(y_test, predicted_knn_uniform, s=20, edgecolor="black",c="darkgreen", label="Knn data uniform")
plt.scatter(y_test, predicted_knn_distance, s=20, edgecolor="black",c="darkblue", label="Knn data distance")
plt.xlabel("Data")
plt.ylabel("Target")
plt.legend()
plt.show()
```

```python
best_RMS = 1000
best_nn = 1
A = np.zeros(len(y_train), dtype=np.float)
for nn in np.arange(1, len(y_train) + 1):
    clf_knn = neighbors.KNeighborsRegressor(nn, weights='distance')
    clf_knn.fit(X_train, y_train)
    predicted = clf_knn.predict(X_test)
    RMS = np.sqrt(np.mean((predicted - y_test) ** 2))
    index = nn - 1
    A[index] = RMS
    if RMS < best_RMS:
        best_RMS = RMS
        best_nn = nn
    # print("k neighbors = " + str(nn))
    # print("RMS = " + str(RMS))

print("\n")
print("best k = " + str(best_nn))
print("best RMS = " + str(best_RMS))

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.yticks(np.arange(0.0, 1.0, 0.01))

plt.plot(np.arange(1, len(y_train) + 1), A)

plt.show()
```
